# Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

# 1. Load Data

```python
# Load the dataset
df = pd.read_csv('../sample_dataset/student_performance_dataset.csv')
```

```python
# Select only the columns we need for our model
df_model = df[['study_hours_per_day', 'attendance_percentage', 'exam_score']].copy()
```

# 2. Initial Data Inspection

```python
print("First 5 rows:")
print(df_model.head())
```

```python
print("\nDataset Info:")
df_model.info()
```

```python
print("\nStatistical Summary:")
print(df_model.describe())
```

# 3. Visualization

```python
# Create a directory for saving plots if it doesn't exist
os.makedirs('../visualizations', exist_ok=True)
```

```python
# Create a pairplot to see relationships between all variables
sns.pairplot(df_model)
plt.suptitle('Pairplot of Study Hours, Attendance, and Exam Score', y=1.02)
plt.savefig('../visualizations/pairplot.png')
plt.show()
```

```python
# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
correlation_matrix = df_model.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig('../visualizations/correlation_heatmap.png')
plt.show()
```
